```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Average Movie Rating")
sc = SparkContext.getOrCreate(conf = conf)
```

```python
movies = sc.textFile('/movies.csv')
movies.collect()
```

```python
preview_movie_ratings = movies.map(lambda x: x.split('\t'))
preview_movie_ratings.collect()
```

```python
movie_ratings = preview_movie_ratings.map(lambda x: (x[0], (int(x[1]), 1)))
movie_ratings.collect()
```

```python
combined_movie_ratings = movie_ratings.reduceByKey(lambda x,y : (x[0]+y[0], x[1]+y[1]))
combined_movie_ratings.collect()
```

```python
avg_movie_ratings = combined_movie_ratings.map(lambda x : (x[0] , x[1][0]/x[1][1]))
avg_movie_ratings.collect()
```

```python
avg_movie_ratings.saveAsTextFile('/avg_ratings.csv')
```
